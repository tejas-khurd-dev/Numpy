ADVANCE ARRAY MANIPULATION

Stacking of array

```python
# You can't stack matrix
```

```python
s1 = np.array([1, 2, 3, 4])
s2 = np.array([5, 6, 7, 8])
```

```python
np.vstack((s1,s2)) # vertical (row) stacking
```

```python
np.hstack((s1,s2)) # horizontal (column) stacking
```

```python
np.column_stack((s1,s2)) # take 1 ele from s1 abd 1 ele from s2 and make one row follwed by others
```

Splitting array

```python
c = np.arange(16).reshape(4, 4)
c
```

```python
np.hsplit(c,2) 
```

```python
arr_split = np.vsplit(c,4)
arr_split
```

```python
for i in arr_split:
    print(i)
```
