```python
import numpy as np
```

Indexing and slicing of vectors

```python
arr = np.arange(61,71)
arr
```

Indexing

```python
arr[6]
```

Slicing

```python
arr[2:7:2] # arr[start: stop-1: steps] => by default it is arr[0: len(arr): 1]
```

```python
arr[:7:]
```

```python
arr[7:]
```

Indexing and slicing of matrix

```python
arr2 = np.arange(61,81)
arr2
```

```python
arr2.size
```

```python
arr2 = arr2.reshape(5,4)
arr2
```

Indexing

```python
arr2[2] # print row
```

```python
arr2[0,0] 
```

Slicing

```python
arr2[0:2,1:3] # arr[start: stop-1, start: stop-1] or arr[row_num, start:stop-1 ]
```

```python
arr2[4, 2:] # or arr2[4:5, 2:4]
```

```python
arr2[:, 2]  # slice columns
```

Boolean Indexing

```python
arr3 = np.arange(1,11)
arr3
```

```python
bool_index = arr3%2 ==0
bool_index
```

```python
arr4 = arr3[bool_index] # arr4 mai arr3 ke jo true values hai woh aajayenge 
arr4
```
